```python
import matplotlib.pyplot as plt
```

```python
ages = [32, 40, 38, 45, 50, 34, 42, 36, 48, 55,
        39, 41, 37, 44, 49, 43, 35, 52, 46, 40,
        33, 47, 39, 51, 38, 44, 42, 36, 53, 41,
        37, 45, 39, 54, 40, 38, 36, 43, 50, 42,
        38, 55, 39, 37, 49, 46, 44, 40, 35, 52]
plt.figure(figsize=(10,6))
plt.hist(ages, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histogram of Ages of Fitness Program Participants")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
class_intervals = [0, 40, 50, 60, 70, 100]
frequencies = [5, 6, 8, 4, 2]

class_widths = [class_intervals[i+1] - class_intervals[i] for i in range(len(frequencies))]
densities = [frequencies[i] / class_widths[i] for i in range(len(frequencies))]
plt.bar(class_intervals[:-1], densities, width=class_widths, align="edge", edgecolor="black", alpha=0.7)
plt.title("Density Histogram of Weights")
plt.xlabel("Weight (grams)")
plt.ylabel("Density (Frequency / Class Width)")
plt.show()
```

```python
pi_digits = "14159265358979323846264338327950288419716939937510"
digits = [int(digit) for digit in pi_digits]

frequency = [digits.count(i) for i in range(10)]
plt.figure(figsize=(10, 6))
plt.bar(range(10), frequency, color='skyblue', edgecolor='black', alpha=0.7)
plt.xticks(range(10))
plt.title("Frequency Distribution of Digits in π (up to 50 Decimal Places)")
plt.xlabel("Digits (0-9)")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Frequency Distribution of Digits:")
for i in range(10):
    print(f"Digit {i}: {frequency[i]} occurrences")
```
